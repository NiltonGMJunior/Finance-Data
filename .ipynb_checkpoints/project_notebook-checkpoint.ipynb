```python
bank_stocks = pd.read_pickle('all_banks')
```

```python
bank_stocks
```

```python
tickers = ['BAC', 'C', 'GS', 'JPM', 'MS', 'WFC']
```

```python
bank_stocks.xs(key='Close', axis=1, level='Stock Info').max()
```

```python
returns = pd.DataFrame()
```

```python
for tick in tickers:
    returns[tick + ' Return'] = bank_stocks[tick]['Close'].pct_change()
```

```python
returns
```

```python
sns.pairplot(returns[1:])
```

```python
returns.idxmin()
```

```python
returns.idxmax()
```

```python
returns.std()
```

```python
returns.loc['2015-01-01':'2015-12-31'].std()
```

```python
sns.distplot(returns.loc['2015-01-01':'2015-12-31']['MS Return'], color='green', bins=50)
```

```python
sns.distplot(returns.loc['2008-01-01':'2008-12-31']['C Return'],color='red',bins=50)
```

```python
import matplotlib.pyplot as plt
sns.set_style('whitegrid')
```

```python
for tick in tickers:
    bank_stocks[tick]['Close'].plot(label=tick,figsize=(12,4))
plt.legend()
```

```python
bank_stocks.xs(key='Close',axis=1,level='Stock Info').plot(figsize=(12,4))
```

```python
plt.figure(figsize=(12,4))
bank_stocks['BAC']['Close'].loc['2008-01-01':'2009-01-01'].rolling(window=30).mean().plot(label='30 day moving average for BAC stock closing price')
bank_stocks['BAC']['Close'].loc['2008-01-01':'2009-01-01'].plot(label='BAC stock closing price')
plt.legend()
```

```python
sns.heatmap(bank_stocks.xs(key='Close',axis=1,level='Stock Info').corr(),annot=True,cmap='coolwarm')
```
